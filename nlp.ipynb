```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
from pylab import rcParams
rcParams['figure.figsize'] = 20, 14
```

```python
Mbti = pd.read_csv('mbti_1.csv')
Mbti
```

```python
first_plot = Mbti.groupby('type').agg({'type':'count'})
firstplot = first_plot.rename(columns={'type':'counts'}).reset_index()
```

```python
sns.barplot('type','counts',data= firstplot)
```

```python
##plot 2 
from scipy.misc import imread
from wordcloud import WordCloud,STOPWORDS

fig, ax = plt.subplots(len(Mbti['type'].unique()), figsize=(24,14*10))

for i in range(16):
    typeplot= Mbti[Mbti['type']==Mbti['type'].unique()[i]]
    bigband = WordCloud().generate(typeplot.to_string())
    ax[i].imshow(bigband)
    ax[i].axis('off')
```
